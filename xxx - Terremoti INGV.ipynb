```python
import pandas as pd
from datetime import datetime  
df = pd.read_csv('data/dati_fle_2010_2019.csv', header=None, names=['Data', 'Magnitudo','Latitudine','Longitudine','Profondità'])
df.drop(0, inplace=True)
for col in ['Magnitudo', 'Latitudine', 'Longitudine']:
    df[col] = pd.to_numeric(df[col])
df['Time'] = pd.to_datetime(df.Data)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,6), dpi=120)
plt.scatter(df.Time, df.Magnitudo, s=4, alpha=0.2)
plt.xlabel('Data')
plt.ylabel('Magnitudo')
plt.show()
```

```python
plt.figure(figsize=(12,3), dpi=120)
plt.scatter(df.Time, df['Latitudine'], s=4, alpha=0.2)
plt.xlabel('Data')
plt.ylabel('Latitudine')
plt.show()
plt.figure(figsize=(12,3), dpi=120)
plt.scatter(df.Time, df['Longitudine'], s=4, alpha=0.2)
plt.xlabel('Data')
plt.ylabel('Longitudine')
plt.show()
```

```python
plt.figure(figsize=(12,6), dpi=120)
plt.scatter(df['Longitudine'], df['Latitudine'], s=(df.Magnitudo+1)*10, alpha=0.2)
plt.xlabel('Longitudine')
plt.ylabel('Latitudine')
plt.show()
```

```python
lon_mean, lat_mean = df['Longitudine'].mean(),df['Latitudine'].mean()
plt.figure(figsize=(8,8), dpi=120)
plt.scatter(df['Longitudine']-lon_mean, df['Latitudine']-lat_mean, s=(df.Magnitudo+1)*20, alpha=0.2)
plt.xlabel('Longitudine')
plt.ylabel('Latitudine')
plt.xlim(-0.04, 0.04)
plt.ylim(-0.04, 0.04)
plt.show()
```

```python
plt.figure(figsize=(12,6), dpi=120)
plt.scatter(((df['Longitudine']-lon_mean)**2+(df['Latitudine']-lat_mean)**2)**0.5, df.Magnitudo, s=10, alpha=0.1)
plt.xlabel('Distanza')
plt.ylabel('Magnitudo')
plt.show()
```

```python
plt.figure(figsize=(12,6), dpi=120)
plt.scatter(df.Time, ((df['Longitudine']-lon_mean)**2+(df['Latitudine']-lat_mean)**2)**0.5, s=10, alpha=0.1)
plt.xlabel('Data')
plt.ylabel('Distanza')
plt.ylim(0,0.02)
plt.show()
```
